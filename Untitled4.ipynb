# Step 0: Get tweets from specific users

```python
#grab tweets,likes, time, and users specified (number_of_tweets is how many tweets we want to grab)
# function to store tweets,likes,time, users 
def grab_tweets_users(number_of_tweets,list_of_users_to_follow):
#    number_of_tweets = 100
    tweets = []
    likes = []
    time = []
    user=[]
#    superfollows=['JonahLupton','mukund','cperruna','saxena_puru']
    for x in range(len(list_of_users_to_follow)):
        for i in tweepy.Cursor(api.user_timeline, id=list_of_users_to_follow[x], tweet_mode='extended').items(number_of_tweets):
            tweets.append(i.full_text)
            likes.append(i.favorite_count)
            time.append(i.created_at)
            user.append(list_of_users_to_follow[x])
    
    df = pd.DataFrame({'tweets':tweets,'likes':likes,'time':time,'user':user})
    #remove retweets from dataframe
    df=df[df['tweets'].str.contains('RT ')==False]

    return df    

#example below how to pull        
superfollows=['JonahLupton','mukund','cperruna','saxena_puru','richard_chu97','Crussian17','BahamaBen9','ChaseGreatness','deadnsyde','dhaval_kotecha','jaminball','Beth_Kindig']
data=grab_tweets_users(100,superfollows)
display(data.tail())
```

# Step 1: Create sentiment counting function and create a new column by date

```python
#here we will pull textblob to create sentiment and graph the chart by days and sentiment (might need to install TextBlob library for sentiment)
data=data[data['time']>'2021-05-30 09:06:40	'] #storing data from certain date and onward
data['hour']=data['time'].dt.floor('d') #floring date by date 
display(data.head())

# function to store sentiment of each tweet
def blob_portfolio(info):
    info['blob'] = info['tweets'].apply(lambda x: TextBlob(x).polarity)
    plt.figure(figsize=[15,10])

    info.groupby('hour')['blob'].mean().plot(kind='bar',color='red')
    plt.title('Sentiment of tweets from my top follows in twitter')
    plt.ylabel('Sentiment between -1 to +1')
    plt.show()

blob_portfolio(data)
```

```python
# function to create new column ticker and store words that follow $ since in twitter people denote stocks by ex: $TSLA
def find_ticker(text):
    ticker=re.findall(r'(\$[^.,;\s\d]+)',text)
    return " ".join(ticker)

data['ticker']=data['tweets'].apply(lambda x: find_ticker(x))



#display(data['tweets'][14])
display(data.head(7))
#display(data['ticker'].unique())#['timeline_of_user'])

```
